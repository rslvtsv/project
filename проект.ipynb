### Описание данных

Файл /datasets/rest_info.csv содержит информацию о заведениях общественного питания:

name — название заведения;

address — адрес заведения;

district — административный район, в котором находится заведение, например Центральный административный округ;

category — категория заведения, например «кафе», «пиццерия» или «кофейня»;

hours — информация о днях и часах работы;

rating — рейтинг заведения по оценкам пользователей в Яндекс Картах (высшая оценка — 5.0);

chain — число, выраженное 0 или 1, которое показывает, является ли заведение сетевым (для маленьких сетей могут встречаться ошибки): 
0 — заведение не является сетевым;
1 — заведение является сетевым.
seats — количество посадочных мест.

Файл /datasets/rest_price.csv содержит информацию о среднем чеке в заведениях общественного питания:

price — категория цен в заведении, например «средние», «ниже среднего», «выше среднего» и так далее;

avg_bill — строка, которая хранит среднюю стоимость заказа в виде диапазона, например: 
«Средний счёт: 1000–1500 ₽»;
«Цена чашки капучино: 130–220 ₽»;
«Цена бокала пива: 400–600 ₽».
и так далее;

middle_avg_bill — число с оценкой среднего чека, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Средний счёт»: 
Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
Если значения нет или оно не начинается с подстроки «Средний счёт», то в столбец ничего не войдёт.

middle_coffee_cup — число с оценкой одной чашки капучино, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Цена одной чашки капучино»: 
Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
Если значения нет или оно не начинается с подстроки «Цена одной чашки капучино», то в столбец ничего не войдёт.

## 2. Предобработка данных

```python
df['seats'] = df['seats'].fillna(-1)
df['seats'] = df['seats'].astype('int64')
```

```python
df['price'] = df['price'].fillna(-1)
df['avg_bill'] = df['avg_bill'].fillna(-1)
df['middle_avg_bill'] = df['middle_avg_bill'].fillna(-1)
df['middle_coffee_cup'] = df['middle_coffee_cup'].fillna(-1)
df.isna().sum()
```

```python
duplicates = df[df.duplicated(subset=['name', 'address'], keep=False)]
print(duplicates.sort_values(by=['name', 'address']))
```

Вывели дубликаты

```python
df = df.drop_duplicates(subset=['name', 'address'], keep='first')
```

### Задача 3

Изучите соотношение сетевых и несетевых заведений в целом по всем данным и в разрезе категорий заведения. Каких заведений больше — сетевых или несетевых? Какие категории заведений чаще являются сетевыми? Исследуйте данные, ответьте на вопросы и постройте необходимые визуализации.

```python
chain_counts = df['chain'].value_counts()
print("Соотношение сетевых и несетевых заведений:")
print(chain_counts)

# Простая круговая диаграмма
chain_counts.plot.pie(
    autopct='%1.1f%%',
    labels=['Несетевые', 'Сетевые'],
    colors=['lightcoral', 'lightgreen'],
    title='Соотношение сетевых и несетевых заведений'
)
plt.ylabel('')
plt.show()
```

Доля несетевых зведений больше на 20%. Это может быть связано с тем, что большинство маленьких заведений так и не перерастают в сетевые. 

```python
category_chain = df.groupby(['category', 'chain']).size().unstack()
category_chain.columns = ['Несетевые', 'Сетевые']
category_chain['Доля сетевых'] = category_chain['Сетевые'] / (category_chain['Несетевые'] + category_chain['Сетевые'])
category_chain.sort_values('Доля сетевых', ascending=False)['Доля сетевых'].plot.bar(
    figsize=(12, 6),
    color='lightblue',
    title='Доля сетевых заведений по категориям'
)
plt.xlabel('Категория')
plt.ylabel('Доля сетевых заведений')
plt.xticks(rotation=45)
plt.show()
```

Чаще всего сетевыми заведениями становятся булочные - обычно они расположены в очень проходимых местах: рядом с офисами, торговыми центрами и ВУЗами.

### Задача 6

Изучите, с какими данными показывают самую сильную корреляцию рейтинги заведений? Постройте и визуализируйте матрицу корреляции рейтинга заведения с разными данными: его категория, положение (административный район Москвы), статус сетевого заведения, количество мест, ценовая категория и признак, является ли заведения круглосуточным. Выберите самую сильную связь и проверьте её.

```python
corr_data = df[['rating', 'chain', 'seats', 'is_24_7']].copy()


corr_data['price'] = df['price'].astype('category').cat.codes  
corr_data['district'] = df['district'].astype('category').cat.codes 
corr_data['category'] = df['category'].astype('category').cat.codes  

corr_matrix = corr_data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix[['rating']].sort_values('rating', ascending=False),
            annot=True, cmap='coolwarm', center=0,
            fmt='.2f', linewidths=0.5)
plt.title('Корреляция рейтинга с другими показателями', pad=20)
plt.tight_layout()
plt.show()

strongest_corr = corr_matrix['rating'].abs().sort_values(ascending=False).index[1]
print(f"\nСамая сильная корреляция: рейтинг и {strongest_corr}")

if strongest_corr in ['seats', 'price']:
    sns.lmplot(x=strongest_corr, y='rating', data=df, height=6, aspect=1.5)
else:
    sns.boxplot(x=strongest_corr, y='rating', data=df)
plt.title(f'Зависимость рейтинга от {strongest_corr}')
plt.show()
```

Оказалось, что из всех изученных факторов - ценовой категории, района, типа заведения, количества мест и сетевого статуса - только круглосуточная работа хоть как-то влияет на рейтинг, да и то совсем немного (корреляция всего 0.12). Это видно по тому, что у заведений с is_24_7=True средний рейтинг чуть выше. Все остальные параметры практически не коррелируют с оценками посетителей - коэффициенты мизерные (0.02-0.03). Особенно неожиданно, что ни цена, ни категория заведения (кафе, ресторан и т.д.) не показывают значимой связи с рейтингом. Получается, что посетители ставят оценки, не особо ориентируясь на эти характеристики. Такие слабые корреляции наводят на мысль, что главные факторы, влияющие на рейтинг - это что-то другое, чего нет в нашем наборе данных. Вероятно, качество обслуживания, вкус еды или атмосфера заведения играют куда большую роль, чем формальные параметры вроде количества мест или принадлежности к сети. Круглосуточные заведения, видимо, получают небольшой "бонус" за удобство, но в целом картина выходит довольно равномерной.

### Задача 7

Сгруппируйте данные по названиям заведений и найдите топ-15 популярных сетей в Москве. Для них посчитайте значения среднего рейтинга. Под популярностью понимается количество заведений этой сети в регионе. К какой категории заведений они относятся? Результат сопроводите подходящими визуализациями.

```python
chains = df[df['chain'] == 1].copy()

top_chains = chains['name'].value_counts().head(15)

chain_stats = chains.groupby('name').agg(
    count=('name', 'size'),
    avg_rating=('rating', 'mean'),
    category=('category', lambda x: x.mode()[0]) 
).sort_values('count', ascending=False).head(15)

plt.figure(figsize=(12, 6))
sns.barplot(x=chain_stats['count'], y=chain_stats.index, palette='viridis')
plt.title('Топ-15 сетей общепита Москвы по количеству заведений', fontsize=14)
plt.xlabel('Количество заведений')
plt.ylabel('Название сети')
plt.show()
```

В топ-5 также входят кофейни ("Кофе Хауз") и пекарни ("Булошная"). В целом видно три основных типа сетей:
1. Кофейни (4-5 позиций в топ-15)

При этом разрыв между первым и последним местом значительный — у "Шоколадницы" в 2-3 раза больше точек, чем у замыкающих топ-15 сетей. Это говорит о высокой концентрации рынка — несколько крупных игроков доминируют, а остальные существенно уступают по охвату.

### Задача 8

Изучите вариацию среднего чека заведения (столбец `middle_avg_bill`) в зависимости от района Москвы. Проанализируйте цены в Центральном административном округе и других. Как удалённость от центра влияет на цены в заведениях? Результат сопроводите подходящими визуализациями.

```python

# 1. Подготовка данных
# Преобразуем средний чек в число
df['middle_avg_bill'] = pd.to_numeric(df['middle_avg_bill'], errors='coerce')

# Группировка для ЦАО vs остальные
center_avg = df[df['district'] == 'Центральный административный округ']['middle_avg_bill'].mean()
other_avg = df[df['district'] != 'Центральный административный округ']['middle_avg_bill'].mean()

# 2. Сравнение ЦАО с другими округами (горизонтальная столбчатая диаграмма)
plt.figure(figsize=(10, 4))
sns.barplot(
    x=[center_avg, other_avg],
    y=['Центральный округ', 'Другие районы'],
    palette=['#ff7f0e', '#1f77b4'],
    orient='h'
)
plt.title('Средний чек: ЦАО vs остальные районы')
plt.xlabel('Средний чек (руб.)')
plt.ylabel('')
plt.show()

# 3. Топ-15 районов по среднему чеку (горизонтальная столбчатая диаграмма)
district_means = df.groupby('district')['middle_avg_bill'].mean().sort_values(ascending=False).head(15)

plt.figure(figsize=(12, 6))
sns.barplot(
    x=district_means.values,
    y=district_means.index,
    palette='viridis'
)
plt.title('Топ-15 районов по среднему чеку')
plt.xlabel('Средний чек (руб.)')
plt.ylabel('Район')
plt.show()
```

Анализ графиков показывает, что средний чек в Центральном административном округе (более 500 рублей) значительно превышает показатели остальных округов. Наиболее высокие значения среднего чека наблюдаются в районах, близких к центру. Юго-восточный (менее 200 рублей) и северо-восточный округа (около 250 рублей) демонстрируют относительно низкие показатели среднего чека. Таким образом, местоположение оказывает существенное влияние на величину среднего чека.
